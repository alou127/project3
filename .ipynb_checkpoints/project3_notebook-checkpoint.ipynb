```python
# Make Education dummies
edu_names = ['edu_someHS','edu_HS','edu_someCollege','edu_college','edu_nan']
edu_dummy = pd.get_dummies(df['education'], dummy_na=True)
edu_dummy.head()
```

```python
for i in range(4):
    edu_dummy.rename({i+1:edu_names[i]},axis=1,inplace=True)

edu_dummy.rename({'nan':'edu_nan'},axis=1,inplace=True)
```

```python
#check to see if ppl have entered double lvls of education
edu_dummy_T = edu_dummy.T
for i in edu_dummy_T.columns:
    if sum(edu_dummy_T[i]) != 1:
        print(i)
```

```python
df = pd.concat([df, edu_dummy], axis=1)
```

```python
df.drop('education',axis=1,inplace=True)
df.head().T
```

# Classification Models

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X_cols = ['male', 'age', 'currentSmoker', 'cigsPerDay', 'BPMeds',
       'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
       'diaBP', 'BMI', 'heartRate', 'glucose', 'edu_someHS',
       'edu_HS', 'edu_someCollege', 'edu_college', 'nan']
X_cols_trial = ['male', 'cigsPerDay', 'BPMeds',
       'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
       'diaBP', 'BMI', 'heartRate', 'glucose']
```

```python
y = df_imputed_mask.TenYearCHD
X = df_imputed_mask[X_cols]
y_trial = df_imputed_mask.TenYearCHD
X_trial = df_imputed_mask[X_cols_trial]
```

### Test/Val/Train split

```python
from sklearn.model_selection import train_test_split
```

```python
# hold out 20% of the data for final testing
X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=555)
X_trial, X_test_trial, y_trial, y_test_trial = train_test_split(X_trial, y_trial, test_size=.2, random_state=555)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=.25, random_state=33)
X_train_trial, X_val_trial, y_train_trial, y_val_trial = train_test_split(X_trial, y_trial, test_size=.25, random_state=33)
```

```python
def print_metrics(model,X_set,y_set):
    y_model_pred = model.predict(X_set)
    print(metrics.classification_report(y_set, y_model_pred))
```

### Random Forest

```python
RC = RandomForestClassifier(n_estimators=100, bootstrap=True)
```

```python
RC.fit(X_train, y_train)
```

```python
y_pred_val = RC.predict(X_val)
```

```python
print_metrics(RC,X_val,y_val)
```

```python
print_metrics(RC,X_train,y_train)
```

### Let's smote the class imbalance

```python
from imblearn.over_sampling import SMOTE
X_smoted, y_smoted = SMOTE(random_state=42).fit_sample(X_train,y_train)
```

```python
RC_smote = RandomForestClassifier(n_estimators=1000, max_features='auto', max_depth=3, random_state=17)
RC_smote.fit(X_smoted, y_smoted)

print_metrics(RC_smote,X_val,y_val)
```

### RF with SMOTE

```python
RC_smote_cb = RandomForestClassifier(n_estimators=1000, max_features=10, max_depth=4, 
                                     random_state=17, class_weight=None)
RC_smote_cb.fit(X_smoted, y_smoted)

print_metrics(RC_smote_cb,X_val,y_val)
```
